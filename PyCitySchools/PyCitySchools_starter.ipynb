```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

```python
#District Summary
school_count = school_data["school_name"].count()
student_count = school_data_complete["student_name"].count()
total_budget = '${:,.2f}'.format(school_data["budget"].sum())

avg_math = round(school_data_complete["math_score"].mean(),2)
avg_read = round(school_data_complete["reading_score"].mean(),2)

pass_math = school_data_complete[school_data_complete["math_score"]>=70]
pct_pass_math = round(pass_math.count()["student_name"]/student_count*100,2)

pass_read = school_data_complete[school_data_complete["reading_score"]>=70]
pct_pass_read = round(pass_read.count()["student_name"]/student_count*100,2)

pass_overall = school_data_complete[(school_data_complete["reading_score"]>=70)&(school_data_complete["math_score"]>=70)]
pct_pass_overall = round(pass_overall.count()["student_name"]/student_count*100,2)

result ={"Total Schools":[school_count], 
         "Total Students":[student_count], 
         "Total Budget":[total_budget],
         "Average Math Score": [avg_math], 
         "Average Reading Score": [avg_read], 
         "% Passing Math": [pct_pass_math], 
         "% Passing Reading": [pct_pass_read], 
         "% Overall Passing": [pct_pass_overall]}

result_df = pd.DataFrame(result)
result_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Summary
group_school_name = school_data_complete.groupby(["school_name"])
school_type = group_school_name["type"].first()
total_student = group_school_name.size()
total_budget = group_school_name["budget"].first()
per_stu_budget = total_budget/total_student

avg_math = round(group_school_name["math_score"].mean(),2)
avg_read = round(group_school_name["reading_score"].mean(),2)

pass_math = school_data_complete[school_data_complete["math_score"]>=70].groupby(["school_name"])
pct_pass_math = round(pass_math.count()["student_name"]/total_student*100,2)

pass_read = school_data_complete[school_data_complete["reading_score"]>=70].groupby(["school_name"])
pct_pass_read = round(pass_read.count()["student_name"]/total_student*100,2)

pass_overall = school_data_complete[(school_data_complete["reading_score"]>=70)&(school_data_complete["math_score"]>=70)].groupby(["school_name"])
pct_pass_overall = round(pass_overall.count()["student_name"]/total_student*100,2)




summary ={"School Type": school_type,
        "Total Students": total_student,
        "Total School Budget": total_budget,
        "Per Student Budget": per_stu_budget,
        "Average Math Score": avg_math,
        "Average Reading Score": avg_read,
        "% Passing Math":pct_pass_math,
        "% Passing Reading" : pct_pass_read,
        "% Overall Passing" : pct_pass_overall
         }

summary_df = pd.DataFrame(summary)
summary_df
```

## Top Performing Schools (By % Overall Passing)

```python

summary_df.sort_values(by=["% Overall Passing"], ascending=False, inplace = True)
summary_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#creating bins for per student budget
bins = [0, 584,629,644,675]
group_label_budget = ["<$584", "$585-629", "$630-644", "$645-675"]


summary_df["Spending Ranges"] = pd.cut(summary_df["Per Student Budget"], bins, labels=group_label_budget)
summary_df



spending_group = summary_df.groupby(["Spending Ranges"])

round(spending_group[["Average Math Score",
        "Average Reading Score",
        "% Passing Math",
        "% Passing Reading", 
        "% Overall Passing"]].mean(),2)

```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#creating bins for school size
bins = [0, 1000, 2000, 5000]
group_label = ["Small(<1000)", "Medium(1000-2000)", "Large(2000-3000)"]


summary_df["School Size"] = pd.cut(summary_df["Total Students"], bins, labels=group_label)
summary_df



school_size = summary_df.groupby(["School Size"])

round(school_size[["Average Math Score",
        "Average Reading Score",
        "% Passing Math",
        "% Passing Reading", 
        "% Overall Passing"]].mean(),2)
```
